```python
# Required Header
%matplotlib inline

import matplotlib.pyplot as plt
import networkx as nx
```

## Get a edge list data set
### 1. Download a edge list that has more than 1000 nodes. Careful not to pick very large graphs
### 2. Store the edge list as a graph call it it G1

```python
el = open ("myedge.txt", 'rb') # open file and rb is for read-binary
G1 = nx.read_edgelist(el, nodetype=int) # edges on each line of the input
el.close() # we can close the file now
```

## Plot the graph G1 without node numbers

```python
nx.draw_circular(G1, node_color='C0',node_size=100,with_labels=False)
plt.show()
```

## Find the degree of each node in G1 and draw a histogram: nodes on the x-axis and degree on the y-axis

```python
degree_sequence = [d for n, d in G1.degree()]
plt.hist(x=degree_sequence, bins='auto', color='#0504aa',
                            alpha=0.7, rwidth=0.85,orientation='horizontal')
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Number of Nodes')
plt.ylabel('Degree')
plt.show()
```

## Find the average degree of nodes in G1

```python
average=sum(degree_sequence)/len(degree_sequence)
print("The average degree of nodes is", average)
```

## Construct a subgraph G2 from G1 by removing nodes whose degree is less than 2 times the average degree and nodes whose degree is greater than 2 times the avearge degree
I guess the real meaning is to remove nodes whose degree < 1/2 average degree and >2 avergage degree

```python
G2_nodes=[]
G1_degree=dict(G1.degree())

G1_degree.keys()
count=0
for i in G1_degree.keys():
    if degree_sequence[count] >=1/2*average and degree_sequence[count]<=2*average:
            G2_nodes.append(i)
    count=1+count


G2=G1.subgraph(G2_nodes)
```

## Plot the subgraph G2 without node numbers

```python
nx.draw_circular(G2, node_color='C0',node_size=100,with_labels=False)
plt.show()
```

## Create erdos_renyni graph with same number of nodes as in G2
